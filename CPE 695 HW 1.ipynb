Begninning with the provided code snippet below to generate the data

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 10 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

Defining plot function

```python
def plot_polynomial(x, y, m):
    coefficients = np.polyfit(x.flatten(), y.flatten(), m)
    polynomial = np.poly1d(coefficients)

    # Plot the noisy data
    plt.plot(x, y, 'ro', label='Noisy data')
    
    # Generate x values for smooth curve
    x_values = np.linspace(min(x), max(x), 100)
    
    # Plot polynomial curve
    plt.plot(x_values, polynomial(x_values), label=f'Polynomial (m={m})')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Polynomial fitting with m={m}')
    plt.show()
    
    # Calculate and return MSE
    y_prediction = polynomial(x)
    mse = calculate_mse(y, y_prediction)
    return mse

# Function to calculate MSE
def calculate_mse(y_true, y_prediction):
    return np.mean((y_true - y_prediction) ** 2)
```

Plotting the noisy data and the polynomial with the selected m value of 3

```python
m = 3
plot_polynomial(x, y, m)
```

Plotting the noisy data and polynomial for each value of m from 1 to 8

```python
mse_values = []
orders = range(1, 9)
for m in orders:
    mse = plot_polynomial(x, y, m)
    mse_values.append(mse)
```

Plotting the MSE versus order and determining the best value of m

```python
plt.plot(orders, mse_values, marker='o')
plt.xlabel('Order (m)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE versus order (m)')
plt.xticks(orders)
plt.grid(True)
plt.show()
print(f"{mse_values}")
```

```python
best_m = np.argmin(mse_values) + 1
print(f"Best choice of m: {best_m}")
```
